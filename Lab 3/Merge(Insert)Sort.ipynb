## Defining MergeSort and Merge functions

```
def MergeSort(ar, start, end):
  if(start<end):
    mid = (start+end)/2
    mid = int(mid)
    MergeSort(ar, start, mid)
    MergeSort(ar, mid+1, end)
    Merge(ar, start, mid, end)

def Merge(ar, start, mid, end):
  p = start
  q = mid+1
  temp = []
  for i in range(0,end-start+1):
    if p>mid :
      temp.append(ar[q])
      q += 1
    elif q>end :
      temp.append(ar[p])
      p += 1
    elif ar[p] > ar[q]:
      temp.append(ar[q])
      q += 1
    else:
      temp.append(ar[p])
      p += 1
  i = 0
  for q in range(start, end+1):
    ar[q] = temp[i]
    i += 1
    
    
def insertion_sort(collection, start, end):
    for loop_index in range(start, end+1):
        insertion_index = loop_index
        while insertion_index > 0 and collection[insertion_index - 1] > collection[insertion_index]:
            collection[insertion_index], collection[insertion_index - 1] = collection[insertion_index - 1], collection[insertion_index]
            insertion_index -= 1
```

## Use this part to find the largest size 'S' for which insertion sort is faster than merge sort. We will pass that 'S' value to our Merge_insert_sort function.

```
yeet = []
huha = []
for i in range(10):
  x = random.randint(0,100)
  yeet.append(x)
  huha.append(x)
  
start_time = time.time()
MergeSort(yeet, 0, len(yeet) - 1)
time_taken = time.time() - start_time
print(yeet)
print("Time taken for merge sort = ", time_taken*1000)

start_time = time.time()
insertion_sort(huha, 1, len(huha)-1)
time_taken = time.time() - start_time
print(huha)
print("Time taken in insertion sort = ", time_taken*1000)
```

## Defining Merge(Insert)Sort with a factor 'S'

```
def Merge_insert_Sort(ar, start, end, S):
  if(end-start > S):
    mid = (start+end)/2
    mid = int(mid)
    Merge_insert_Sort(ar, start, mid, S)
    Merge_insert_Sort(ar, mid+1, end, S)
    Merge(ar, start, mid, end)
  else:
    insertion_sort(ar, start, end)

```

```
wtf = []
amigo = []
for i in range(1000):
  y = random.randint(0,10000)
  wtf.append(y)
  amigo.append(y)
  
start_time = time.time()
Merge_insert_Sort(wtf, 0, len(wtf) - 1, 10)
time_taken = time.time() - start_time
print(yeet)
print("Time taken for Merge(Insert)Sort = ", time_taken*1000)

start_time = time.time()
MergeSort(amigo, 0, len(amigo)-1)
time_taken = time.time() - start_time
print(amigo)
print("Time taken for MergeSort = ", time_taken*1000)
```
