## Importing the required packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
csv_data = pd.read_csv('FuelConsumption.csv')
csv_data.head()
```

```python
data = csv_data["CO2EMISSIONS"]
Hash_values = np.asanyarray(data)
```

```python
Hash_values = Hash_values*10
Hash_values
```

```python
hash_table_double1 = [None] * 1069
hash_table_double2 = [None] * 1069
global_size = 1069
```

```python
def h(k):
    return (k % global_size)

def h1(k):
    return ((k % 103) + 3)

def h2(k):
    return ((k % 487) + 7)
```

```python
def H1(k, i):
    return (h(k) + i*h1(k)) % global_size

def H2(k, i):
    return (h(k) + i*h2(k)) % global_size
```

```python
def hash_double1(key):
    i = 0
    while(i != 1068):
        j = H1(key, i)
        if(hash_table_double1[j] == None):
            hash_table_double1[j] = key;
            print(j, key)
            return 
        else:
            i += 1
    print("Overflow")
    
def lookup1():
    val = (int)(input("Enter value to be searched: "))
    print ("\n")
    for number in hash_table_double1:
        if number == val:
            print("Succesful")
            break;
    else:
        print("Number does not exist in the table")
```

```python
for j in range(100):
    hash_double1(Hash_values[j])
```

```python
def hash_double2(key):
    i = 0
    while(i != 1068):
        j = H2(key, i)
        if(hash_table_double2[j] == None):
            hash_table_double2[j] = key;
            print(j, key)
            return 
        else:
            i += 1
    print("Overflow")
    
def lookup2():
    val = (int)(input("Enter value to be searched: "))
    print ("\n")
    for number in hash_table_double2:
        if number == val:
            print("Succesful")
            break;
    else:
        print("Number does not exist in the table")
```

```python
for a in range(100):
    hash_double2(Hash_values[a])
```
